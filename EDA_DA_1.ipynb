```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/smoking.csv"
data = pd.read_csv(url)

# Display the first few rows of the dataset
print(data.head())

# Check columns in the dataset
print("Columns in dataset:", data.columns)

# Get dimensions of the dataset
dimensions = data.shape
print(f"Dimensions: {dimensions}")

# Summary statistics
summary = data.describe(include='all')
print(summary)

# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# Fill or drop missing values as appropriate
data.fillna(method='ffill', inplace=True)  # Example: forward fill

# Check data types
data_types = data.dtypes
print("Data Types:\n", data_types)

# Univariate Analysis: Age Distribution
sns.histplot(data['age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Bivariate Analysis: Age vs Smoking Status
sns.boxplot(x='smoke', y='age', data=data)  # Adjust 'smoking_status' if needed.
plt.title('Age vs  smoke')
plt.xlabel('Smoke')
plt.ylabel('Age')
plt.show()

# Select only numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Compute and display the correlation matrix for numeric features only
correlation_matrix = numeric_data.corr()
print("Correlation Matrix:\n", correlation_matrix)

# Heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Multivariate Analysis: Pair Plot (only for numeric variables)
sns.pairplot(numeric_data)
plt.title('Pair Plot of Numeric Variables')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.cluster import KMeans

# Load the dataset
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/smoking.csv"
data = pd.read_csv(url)

# Display the first few rows of the dataset to verify column names
print("Data preview:\n", data.head())

# Check for duplicates and drop them
data = data.drop_duplicates()
print("Data shape after deduplication:", data.shape)

# Replace missing values for a specific column (e.g., replace NaN values in 'smoke' with 'Non-smoker')
data.fillna(value={'smoke': 'Non-smoker'}, inplace=True)

# Discretization: Bin 'age' into categories
data['age_binned'] = pd.cut(data['age'], bins=5, labels=['very young', 'young', 'middle aged', 'senior', 'elderly'])
print(data[['age', 'age_binned']].head())

# Check for remaining missing values
print("Missing Values:\n", data.isnull().sum())

# Impute missing values with mean for numerical columns and mode for categorical columns
for col in data.columns:
    if data[col].isnull().sum() > 0:  # Check if column has missing values
        if data[col].dtype in ['float64', 'int64']:  # Numerical columns
            data[col] = data[col].fillna(data[col].mean())
        else:  # Categorical columns
            data[col] = data[col].fillna(data[col].mode()[0])

# Univariate Analysis: Histogram for numerical variables
numeric_data = data.select_dtypes(include=['float64', 'int64'])
for col in numeric_data.columns:
    sns.histplot(data[col], kde=True)
    plt.title(f'Univariate Analysis - {col}')
    plt.show()

# Verify columns before Bivariate Analysis
if 'age' in data.columns and 'smoke' in data.columns:
    sns.boxplot(x='smoke', y='age', data=data)
    plt.title("Bivariate Analysis: Age vs Smoking Status")
    plt.xlabel('Smoking Status')
    plt.ylabel('Age')
    plt.show()
else:
    print("Columns 'age' and 'smoke' are required for Bivariate Analysis and are not found in the dataset.")

# Multivariate Analysis: Pair Plot
sns.pairplot(data, hue='smoke')
plt.show()

# Bar plot for 'smoke' value counts
data['smoke'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Smoking Status Count")
plt.xlabel("Smoking Status")
plt.ylabel("Count")
plt.show()

# Standardize the data for PCA (checking columns to ensure they exist)
if 'age' in data.columns:
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(data[['age']].dropna())

    # Perform PCA
    pca = PCA(n_components=1)
    pca_components = pca.fit_transform(X_scaled)
    data['PCA1'] = pca_components[:, 0]

    # PCA Visualization
    sns.scatterplot(x='PCA1', y='age', hue='smoke', data=data)
    plt.title("PCA Results")
    plt.xlabel('PCA1')
    plt.ylabel('Age')
    plt.show()
else:
    print("Column 'age' is required for PCA and is not found in the dataset.")

# Model Development and Evaluation: Linear Regression Model
# Splitting data into training and test sets (ensure 'age' and 'smoke' are available)
if 'age' in data.columns and 'smoke' in data.columns:
    X = data[['age']].dropna()
    y = data['smoke']
    y_encoded = pd.get_dummies(y).idxmax(axis=1).astype('category').cat.codes  # Encode as numerical categories

    X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)

    # Initialize and train the model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Predict and evaluate
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print("Mean Squared Error:", mse)
    print("R^2 Score:", r2)
else:
    print("Columns 'age' and 'smoke' are required for Linear Regression model and are not found in the dataset.")
```
